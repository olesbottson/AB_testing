### –ó–∞–º–µ—Ç–∫–∞ –æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –∫—Ä–∏—Ç–µ—Ä–∏—è—Ö

#### –°—Ü–µ–Ω–∞—Ä–∏–π: —Å—É–¥

–ü—Ä–µ–¥–ø–æ–ª–æ–∂–∏–º, –∏–º–µ–µ—Ç—Å—è –¥–≤–∞ –∫—Ä–∏—Ç–µ—Ä–∏—è:

1) $C_1$ : $\alpha = 0.2$, $\beta=0.2$

2) $C_2$ : $\alpha = 0.05$, $\beta=0.9$

–ü—É—Å—Ç—å –æ–Ω–∏ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è  –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –ø—Ä–∏–Ω–∏–º–∞—Ç—å —Ä–µ—à–µ–Ω–∏—è –æ –≤—ã–Ω–µ—Å–µ–Ω–∏–∏ —Å—É–¥–µ–±–Ω—ã—Ö —Ä–µ—à–µ–Ω–∏–π –≤ –æ—Ç–Ω–æ—à–µ–Ω–∏–∏ –ø–æ–¥–æ–∑—Ä–µ–≤–∞–µ–º—ã—Ö:

$H_0$: —á–µ–ª–æ–≤–µ–∫ –Ω–µ–≤–∏–Ω–æ–≤–µ–Ω

$H_A$: —á–µ–ª–æ–≤–µ–∫ –≤–∏–Ω–æ–≤–µ–Ω

$Error_1 = 10 000 000\$ $ - –ø–æ—Ç–µ—Ä–∏ –æ—Ç —Ç–æ–≥–æ, —á—Ç–æ –±—É–¥–µ—Ç –ø—Ä–∏–≥–æ–≤–æ—Ä–µ–Ω –Ω–µ–≤–∏–Ω–æ–≤–Ω—ã–π —á–µ–ª–æ–≤–µ–∫

$Error_2 = 500 000\$ $ - –ø–æ—Ç–µ—Ä–∏ –æ—Ç —Ç–æ–≥–æ, —á—Ç–æ –ø—Ä–µ—Å—Ç—É–ø–Ω–∏–∫ –±—É–¥–µ—Ç –Ω–∞ —Å–≤–æ–±–æ–¥–µ

–ò –ø—É—Å—Ç—å —Ç–∞–∫–∂–µ –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –¥–æ–ª—è –Ω–µ–≤–∏–Ω–æ–≤–Ω—ã—Ö, –¥–æ—Ö–æ–¥—è—â–∏—Ö –¥–æ —Å—É–¥–∞ —Ä–∞–≤–Ω—è–µ—Ç—Å—è $f_i = 0.2$

–¢–æ–≥–¥–∞ –æ–±—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å –±—É–¥–µ—Ç —Ä–∞–≤–Ω–∞:

$f_i \cdot ( \alpha \cdot Error_1  ) + (1 - f_i) \cdot \beta \cdot Error_2  $


```python
def criteria_cost(alpha, beta, f_i, error1, error2):
  return f_i * alpha * error1 + (1 - f_i) * beta * error2
```

```python
criteria_cost(0.2, 0.2, 0.2, 10**6, 5**5)
```

```python
criteria_cost(0.05, 0.9, 0.2, 10**6, 5**5)
```

–ò–∑ —á–µ–≥–æ –º–æ–∂–Ω–æ –∑–∞–∫–ª—é—á–∏—Ç—å, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –≤—Ç–æ—Ä–æ–≥–æ –∫—Ä–∏—Ç–µ—Ä–∏—è –≥–æ—Ä–∞–∑–¥–æ –±–æ–ª–µ–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ü–µ–ª–∏ –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å

#### –°—Ü–µ–Ω–∞—Ä–∏–π: –∞—ç—Ä–æ–ø–æ—Ä—Ç

–ü—É—Å—Ç—å —Ç–µ–ø–µ—Ä—å —ç—Ç–∏ –∂–µ –∫—Ä–∏—Ç–µ—Ä–∏–∏ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è  –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –ø—Ä–∏–Ω–∏–º–∞—Ç—å —Ä–µ—à–µ–Ω–∏—è –æ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö –ø—Ä–æ–≤–µ—Ä–∫–∞—Ö –≤ –∞—ç—Ä–æ–ø–æ—Ä—Ç—É:

$H_0$: —á–µ–ª–æ–≤–µ–∫ –∑–∞–∫–æ–Ω–æ–ø–æ—Å–ª—É—à–µ–Ω

$H_A$: —á–µ–ª–æ–≤–µ–∫ –Ω–∞–º–µ—Ä–µ–Ω —Å–æ–≤–µ—Ä—à–∏—Ç—å —Ç–µ—Ä–∞–∫—Ç

$Error_1 = 100\$ $ - –ø–æ—Ç–µ—Ä–∏ –æ—Ç —Ç–æ–≥–æ, –æ–±—ã—á–Ω—ã–π —á–µ–ª–æ–≤–µ–∫ –±—É–¥–µ—Ç –ø–æ–¥–≤–µ—Ä–≥–Ω—É—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–π –ø—Ä–æ–≤–µ—Ä–∫–µ

$Error_2 = 1 000 000 000\$ $ - –ø–æ—Ç–µ—Ä–∏ –æ—Ç —Ç–æ–≥–æ, —á—Ç–æ –Ω–∞ –±–æ—Ä—Ç –±—É–¥–µ—Ç –ø—É—â–µ–Ω —Ç–µ—Ä—Ä–æ—Ä–∏—Å—Ç

–ò –ø—É—Å—Ç—å —Ç–∞–∫–∂–µ –∏–∑–≤–µ—Å—Ç–Ω–æ, —á—Ç–æ –¥–æ–ª—è —Ç–µ—Ä—Ä–æ—Ä–∏—Å—Ç–æ–≤ —Å—Ä–µ–¥–∏ –≤—Å–µ—Ö –ø–∞—Å—Å–∞–∂–∏—Ä–æ–≤ —Ä–∞–≤–Ω—è–µ—Ç—Å—è $f_i = 0.00001$

–¢–æ–≥–¥–∞ –æ–±—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø–æ—Ç–µ—Ä—å –±—É–¥–µ—Ç —Ä–∞–≤–Ω–∞:

$f_i \cdot ( \alpha \cdot Error_1  ) + (1 - f_i) \cdot \beta \cdot Error_2  $


```python
def criteria_cost(alpha, beta, f_i, error1, error2):
  return f_i * alpha * error1 + (1 - f_i) * beta * error2
```

```python
c1 = criteria_cost(0.2, 0.2, 0.00001, 100, 10**9)
c2 = criteria_cost(0.05, 0.9, 0.00001, 100, 10**9)

print(c1, c2,  c1 / c2)
```
